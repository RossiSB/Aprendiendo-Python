```python
import pandas as pd
df_viajes = pd.DataFrame({"Viaje": ["Colombia", "Perú", "Australia", "Italia",
                                    "Japón", "Perú", "España", "Portugal"], 
                          "Precio" : [2000, 1800, 2800, 1000, None, 1550, 900, None]})


df_viajes
```

```python
#a: construye un nuevo df donde sólo tenga disponible los viajes de los que me sé el precio.
```

```python
df_n = df_viajes.dropna(subset=["Precio"])
df_new= pd.DataFrame(df_n)
df_new
```

```python
#b: construye una nueva variable llamada Precio_final que se corresponde con el precio inicial aplicado un 5% de descuento.
```
