# Asignaciones y agrupaciones

**Ejercicio 3.8.1 Crea un dataframe,inserta una columna que se llame Dividendos y que sea igual a la columna Beneficio.**

```python
import pandas as pd

datos = {'Mes':['Enero','Febrero','Marzo','Abril','Mayo','Junio','Julio','Agosto','Septiembre','Octubre','Noviembre','Diciembre'],
         'Ventas':[10000, 15200, 14300, 25800, 2560, 12780, 32500, 23624, 12000, 15858, 23250, 11630],
         'Gastos':[6500, 6500, 6500, 7850, 6500, 8250, 6500, 9252, 6500, 10250, 12450, 13000],
         'Beneficio':[3500,8700,7800,17950,-3940,4530,26000,14372,5500,5608,10800,-1370],
         'Cobrado':['S', 'N', 'N', 'S', 'S', 'N', 'S', 'S', 'S', 'N', 'S', 'N']}
df= pd.DataFrame(datos)
df["Dividendo"] = df["Beneficio"]

df
```

**Ejercicio 3.8.2 Utilizando el dataframe anterior, modifica la columna Dividendo para que contenga el 10% de su valor actual utilizando el método assign.**

```python
df1 = df.assign(Dividendo = df["Dividendo"] * 0.1)
df1
```

**Ejercicio 3.8.3 En el resultado obtenido en el ejercicio anterior, verás que hay meses en los que los dividendos son negativos, como esto no puede suceder, modifica la columna Dividendo para que cuando el resultado sea positivo se muestre pero que cuando sea negativo ponga el valor 0. Utiliza la función map.**

```python
def negativo (a):
    if a < 0:
        return 0
    else:
        return a

df1["Dividendo"] = df1["Dividendo"].map(negativo)

df1.loc[:, ["Dividendo"]]


```

**Ejercicio 3.8.4 Multiplica por dos, utilizando la función map, el valor de todos los elementos de la columna Gastos del dataframe.**

```python
def multi (a):
    return a*2
    

df1["Gastos"] = df1["Gastos"].map(multi)

df1

```

```python
dfcopia = df1.copy()
dfcopia
```

**Ejercicio 3.8.5 Recalcula el beneficio y los dividendos mensuales teniendo en cuenta los gastos modificados en el ejercicio anterior.**

```python


df1["Beneficio"] = df1["Ventas"] - df1["Gastos"]
df1

df1["Dividendo"] = df1["Beneficio"] * 0.1
df1
df1["Dividendo"] = df1["Dividendo"].map(negativo)
df1
```

```python
def porcentaje(x):
    return (x["Dividendo"] / x["Ventas"] )*100

df1["Porcentaje_div"] = df1.apply(porcentaje, axis = 1)
df1
```

**Ejercicio 3.8.8 Muestra la media de todos los datos de las columnas numéricas agrupados por si están cobrados o no.**
