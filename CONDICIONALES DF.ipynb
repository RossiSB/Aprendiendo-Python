# Condicionales
Una de las ventajas que nos encontramos a la hora de trabajar con dataframes es que estos
nos ofrecen la posibilidad de aplicar condiciones para filtrar los datos.

**3.7.1 Data**

```python
import pandas as pd
d = pd.read_csv(r"C:\Users\Hp\Downloads\ejecucion-presupuestaria_Chile-febrero-2023.csv", sep=';')
df = pd.DataFrame(d)
df.head()
    
```

**3.7.2 Muestra las columnas Nombre_Subtítulo y Ejecución_de_FEBRERO.**

```python
df[["Nombre_Subtítulo", "Ejecución_de_FEBRERO"]].head()
```

**3.7.3 Muestra los registros cuya ejecución de febrero haya sido menor o igual que 0. Utiliza un condicional simple para realizar la consulta.**

```python
df[df["Ejecución_de_FEBRERO"] <= 0]
```

**3.7.4 Del ejercicio 3.7.3 pero, esta vez, muestra solo el Nombre_Subtítulo y la Ejecución_de_FEBRERO.**

```python
df[df["Ejecución_de_FEBRERO"] <= 0][["Nombre_Subtítulo","Ejecución_de_FEBRERO"]]
```

```python
df.loc[df["Ejecución_de_FEBRERO"]<= 0,["Nombre_Subtítulo","Ejecución_de_FEBRERO"]]
```

**Ejercicio 3.7.5 Utilizando el mismo dataframe, muestra los registros cuya ejecución de febrero sea mayor de 20000 DOLARES**

```python
df.loc[(df["Ejecución_de_FEBRERO"] >20000) & (df["Moneda"] == "DOLARES")]
```

**Ejercicio 3.7.6 Muestra ahora, del mismo dataframe, los Subtítulos y la Ejecución_a_FEBRERO de los registros cuya ejecución a febrero haya sido mayor que 10000000 de PESOS.**

```python
df.loc[(df["Ejecución_a_FEBRERO"] >10000000) & (df["Moneda"] == "PESOS")][["Subtítulo", "Ejecución_a_FEBRERO"]]
```

**Ejercicio 3.7.7 Dada la lista ["IMPUESTOS","RENTAS DE LA PROPIEDAD","INGRESOS DE OPERACIÓN","VENTA DE ACTIVOS FINANCIEROS"] y utilizando el dataframe anterior, muestra los registros cuyo nombre de subtítulo coincida con alguno de los elementos de la lista.**

```python
Lista=["IMPUESTOS","RENTAS DE LA PROPIEDAD","INGRESOS DE OPERACIÓN","VENTA DE ACTIVOS FINANCIEROS"]
df.loc[df["Nombre_Subtítulo"].isin(Lista)]
```

**Ejercicio 3.7.8 Muestra ahora solo los registros cuya ejecución de febrero sea mayor de 10000 DOLARES y cuyo nombre de subtítulo coincida con alguno de los elementos de la lista proporcionada en el ejercicio anterior.**

```python
Lista=["IMPUESTOS","RENTAS DE LA PROPIEDAD","INGRESOS DE OPERACIÓN","VENTA DE ACTIVOS FINANCIEROS"]
df.loc[(df["Ejecución_de_FEBRERO"] > 10000) & (df["Moneda"] == "DOLARES") & (df["Nombre_Subtítulo"].isin(Lista))]
```

**Ejercicio 3.7.9 Utilizando el dataframe creado en el ejercicio 3.7.1, muestra los registros en los que la diferencia entre la Ejecución_a_FEBRERO menos la Ejecución_de_FEBRERO sea igual a 0.**

```python
def diferencia(a,b):
    return(a-b)
df.loc[lambda x: diferencia(x.Ejecución_a_FEBRERO,
x.Ejecución_de_FEBRERO) == 0]
```

```python
def diferencia(a,b):
    return(a-b)
df.loc[lambda x: diferencia(x.Ejecución_a_FEBRERO,
x.Ejecución_de_FEBRERO) == 0][["Nombre_Subtítulo"]]
```
